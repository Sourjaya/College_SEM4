```python
null_var=df.isnull().sum()/df.shape[0]*100
print(null_var)
```

```python
drop_columns=null_var[null_var>17].keys()
drop_columns
```

```python
df1=df.drop(columns=drop_columns)
```

```python
df1.isnull().sum()
```

```python
df1=df1.dropna()
```

```python
df1.shape
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df1.isnull())
```

```python
arr=df1.select_dtypes(include=['int64','float64']).columns
sns.distplot(df['MSSubClass'])
sns.distplot(df1['MSSubClass'])
```

```python
c=1
plt.figure(figsize=(20,35))
for col in arr:
    if col!='Id':
        plt.subplot(9,4,c)
        c=c+1
        sns.distplot(df[col],bins=20)
        sns.distplot(df1[col],bins=20)
```
